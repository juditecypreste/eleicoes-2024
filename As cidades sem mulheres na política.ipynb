## üìÖ 2020

```python
# Definindo as colunas do df
colunas_desejadas = ['SG_UE', 'DS_SIT_TOT_TURNO', 'NM_CANDIDATO', 'NM_URNA_CANDIDATO', 'SG_UF', 'NM_UE', 'SG_PARTIDO', 'DS_GENERO', 'DS_CARGO']

# Lendo o arquivo CSV apenas com as colunas especificadas
df_2000 = pd.read_csv("data/consulta_cand_2000_BRASIL.csv", delimiter=';', encoding='latin-1', usecols=colunas_desejadas)
```

```python
df_2000.head()
```

```python
df_2000 = df_2000[df_2000['DS_SIT_TOT_TURNO'].isin(['ELEITO', 'ELEITO POR M√âDIA'])]
```

```python
mulheres_eleitas = df_2000[(df_2000['DS_GENERO'] == 'FEMININO')]
```

```python
# Obter SG_UE das cidades que elegeram mulheres
sg_ue_com_mulheres = mulheres_eleitas['SG_UE'].unique()
```

```python
# Obter todas as cidades (SG_UF, SG_UE, NM_UE)
cidades_completas = df_2000[['SG_UF', 'SG_UE', 'NM_UE']].drop_duplicates()
```

```python
cidades_sem_mulheres_2000 = cidades_completas[~cidades_completas['SG_UE'].isin(sg_ue_com_mulheres)]
```

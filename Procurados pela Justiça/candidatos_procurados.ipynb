```python
# Importa as bibliotecas necess√°rias para a an√°lise
import pandas as pd
import numpy as np
import re
import numpy as np
import os
```

```python
# L√™ os dados do arquivo CSV contendo informa√ß√µes eleitorais
df = pd.read_csv("data/consulta_cand_2024_BRASIL.csv", delimiter =';', encoding='latin-1')
```

```python
df.head()
```

### üßπ Limpando duplicados

```python
# checando os duplicados segundo o n√∫mero de identifica√ß√£o do t√≠tulo de eleitor
duplicados = df[df.duplicated(subset='NR_TITULO_ELEITORAL_CANDIDATO', keep=False)]
```

```python
df = df.drop_duplicates(subset='NR_TITULO_ELEITORAL_CANDIDATO')
```

```python
df["NR_TITULO_ELEITORAL_CANDIDATO"].nunique()
```

### üóÉÔ∏è Arrumando BNMP

```python
# Pasta com os dfs de mandados por cada estado, com exce√ß√£o de SP que n√£o estava dispon√≠vel para download pelo BNMP na ocasi√£o da reportagem.
mandados = 'data/mandados/sem_sp'
```

```python
# Lista para armazenar os DataFrames
dfs = []

# Percorre todos os arquivos CSV na pasta
for arquivo in os.listdir(mandados):
    if arquivo.endswith('.csv'):
        try:
            # L√™ o CSV ignorando linhas com erro
            le_mandados = pd.read_csv(os.path.join(mandados, arquivo), on_bad_lines='skip', sep=",")
            # Adiciona uma nova coluna com o nome do arquivo (sem a extens√£o)
            le_mandados['ESTADO'] = os.path.splitext(arquivo)[0]
            # Adiciona o DataFrame √† lista
            dfs.append(le_mandados)
        except pd.errors.ParserError as e:
            print(f"Erro ao processar o arquivo {arquivo}: {e}")

# Combina todos os DataFrames em um √∫nico
df_mandados = pd.concat(dfs, ignore_index=True)
```

```python
df_mandados["ESTADO"].unique()
```

### SP

```python
# Base de SP foi conseguida posteriormente e agregada com as demais
sp_mandados = pd.read_csv("data/mandados/SAO PAULO.csv", delimiter =';', encoding='utf-8')
```

```python
sp_mandados.tail()
```

```python
sp_mandados = sp_mandados.drop('Unnamed: 10', axis=1)
```

```python
sp_mandados['ESTADO'] = 'S√ÉO PAULO'
```

```python
df_mandados = pd.concat([df_mandados, sp_mandados], ignore_index=True)
```

```python
df_mandados.tail()
```

```python
df_mandados.shape[0]
```

### üîé Procurando candidatos foragidos

```python
# Padroniza os nomes para min√∫sculas em ambas as tabelas
df['NM_CANDIDATO'] = df['NM_CANDIDATO'].str.upper()
df_mandados['Nome'] = df_mandados['Nome'].str.upper()
```

```python
df['DT_NASCIMENTO'] = pd.to_datetime(df['DT_NASCIMENTO'], dayfirst=True, errors='coerce')
df_mandados['Data de Nascimento'] = pd.to_datetime(df_mandados['Data de Nascimento'], dayfirst=True, errors='coerce')
```

```python
# Fa√ßa a jun√ß√£o das tabelas com base no nome e data de nascimento
tabela_combinada = pd.merge(df, df_mandados, left_on=['NM_CANDIDATO', 'DT_NASCIMENTO'], right_on=['Nome', 'Data de Nascimento'], how='inner')
```

```python
tabela_combinada
```

```python
tabela_combinada.to_csv('data/candidatos_procurados.csv', decimal=";", index=False, encoding='latin-1')
```
